```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from statistics import mean, variance, stdev, mode, median
```

```python

data = pandas.read_csv("C:\\undata.csv")
data.set_index('Country or Area', inplace=True)
data = data.sort_values('Value')
data

for year in range(2010, 2020):
    try:
        data[data["Year"] == year].plot.bar(y='Value',color='g',figsize=(12.5, 20.5), title=year, fontsize=12.5)
    except:
        continue

        
```
